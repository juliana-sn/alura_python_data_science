# Curso: Pandas: Transformação e manipulação de dados

```python
import pandas as pd
```

```python
dados = pd.read_json('/content/dados_hospedagem.json')
dados.head()
```

```python
# Podemos ver que os nossos dados estão em formato de dicionário (dados aninhados), temos que usar o método json_normalize
dados = pd.json_normalize(dados['info_moveis'])
dados.head()
```

```python
# Podemos perceber que algumas colunas têm os dados agrupados tem forma de lista, vamos removê-las (desagrupar)
## Vamos usar o métodoo explode, que separa elementos em linhas

colunas = list(dados.columns)
colunas
```

```python
# Precisamos mexer nas colunas a partir do índice 3
dados = dados.explode(colunas[3:])
dados.head()
```

```python
# Resetar o índice para ordenar os dados
dados.reset_index(inplace=True, drop=True)
# Drop = true para remover os índices antigos

dados.head()
```

```python
dados.info()
```

Podemos perceber que todos os Dtype estão como object e isso está errado

Razões para os dados serem do tipo Object

* Vários tipos de dados diferentes na mesma coluna
* Tipo específico não identificado
* Todos os tipos de dados são str

-----------------------------
Vamos converter os dados numéricos com método astype



```python
# Vamos importar a biblioteca Numpy, que nos fornecerá um tipo de dado (inteiro e float) para podermos converter as colunas númericas
import numpy as np
```

```python
dados['max_hospedes']=dados['max_hospedes'].astype(np.int64)
# Passamos por parâmetro o tipo de dado que queremos que converta
```

```python
dados.info()
```

```python
col_numericas = ['quantidade_banheiros', 'quantidade_quartos', 'quantidade_camas']
```

```python
# Convertendo todas as colunas restantes de tipo int
dados[col_numericas] = dados[col_numericas].astype(np.int64)
```

```python
# Convertendo para float
dados['avaliacao_geral'] = dados['avaliacao_geral'].astype(np.float64)
```

```python
dados.info()
```

Ainda precisamos mexer em algumas colunas como taxas e preços, mas o astype não funciona com símbolos/caracteres.
Vamos tratar essas colunas primeiro

```python
dados['preco']
```

```python
# Vamos usar o método apply com lambda e replace para remover os caracteres

dados['preco'] = dados['preco'].apply(lambda x: x.replace('$', '').replace(',', '').strip())
# strip para apagar espaços em brancos
```

```python
dados['preco'] = dados['preco'].astype(np.float64)
```

```python
dados.info()
```

```python
# Converter as duas colunas das taxas
## Vamos utilizar o método map, e não o apply, pois ele é usado para colunas e linhas em Series, e nesse caso precisamos ler o dataframe
dados[['taxa_deposito', 'taxa_limpeza']] = dados[['taxa_deposito', 'taxa_limpeza']].map(lambda x: x.replace('$', '').replace(',','').strip())
```

```python
dados.head()
```

```python
dados[['taxa_deposito', 'taxa_limpeza']] = dados[['taxa_deposito', 'taxa_limpeza']].astype(np.float64)
```

```python
dados.info()
```

Tratando colunas textuais buscando criar uma estrutura textual mais acessível para realizar uma análise textual. Vamos realizar a manipulação dos textos de cada coluuna textual para criar uma estrutura de tokenização simples
-------------------------------------------------------------

A tokenização é a divisão do texto em tokens, ou seja, unidades menores. Esses tokens podem ser palavras, sílabas, conjuntos de palavras e assim por diante, dependendo do tipo de aplicação.

```python
# Transformando todas as letras em minúsculas
dados['descricao_local'] = dados['descricao_local'].str.lower()
```

```python
dados.head()
```

Removendo caracteres com Regex

Regex é uma expressão regular que se caracteriza por um conjunto de caracteres que fazem uma pesquisa no texto. O Regex é muito utilizado para fazer a busca, validação e manipulação dos dados porque, a partir dessa sequência padronizada de caracteres, conseguimos definir os caracteres que queremos ou não selecionar

```python
dados['descricao_local'][3169]
```

```python
dados['descricao_local'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex=True)
# Regex [^a-zA-Z0-9\-\']
```

```python
dados['descricao_local'] = dados['descricao_local'].str.replace('(?<!\w)-(?!\w)', ' ', regex=True)
```

```python
# Tokenizando a coluna de descrição local
dados['descricao_local'] = dados['descricao_local'].str.split()

dados.head()
```

```python
# Tratando a coluna comodidades
dados['comodidades'] = dados['comodidades'].str.replace('\{|}|\"', '', regex=True)
```

```python
dados['comodidades'] = dados['comodidades'].str.split(',')
dados.head()
```
